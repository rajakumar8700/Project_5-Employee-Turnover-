```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
df_left = df.loc[df["left"] == 1, ["satisfaction_level", "last_evaluation"]]
```

```python
wcss = []
for i in range(1,10):
    model = KMeans(n_clusters=i, random_state=42)
    model.fit(df_left)
    wcss.append(model.inertia_)
plt.plot(range(1,10),wcss)
plt.xlabel('no. of clusters')
plt.ylabel('WCSS')
plt.title('WCSS graph - elbow method')
plt.show()
```

_from the above graph it can be observed that the curve starts flattening at 3 clusters._

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score,cross_val_predict
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

from imblearn.over_sampling import SMOTE
```

```python
dummy = pd.get_dummies(df, drop_first= True, dtype='int')
```

```python
dummy.head()
```

```python
df.tail()
```

```python
X = dummy.drop('left',axis=1)
y = dummy['left']
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 123)
```

```python
smote_dummy = SMOTE(random_state=123)
X_train_smote, y_train_smote = smote_dummy.fit_resample(X_train, y_train)
```

```python
print(f'Distribution before smote : \n {y_train.value_counts()}')
print(f'Distribution after smote : \n {y_train_smote.value_counts()}')
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, confusion_matrix
```

```python
lr = LogisticRegression(max_iter=1000,random_state=42)
lr.fit(X_train_smote,y_train_smote)
kf = KFold(n_splits=5,shuffle=True,random_state=42)
y_train_pred = lr.predict(X_train_smote)
y_test_pred = lr.predict(X_test)
```

```python
y_pred = cross_val_predict(lr,X_train_smote,y_train_smote,cv=kf)
cv_score = cross_val_score(lr,X_train_smote,y_train_smote, scoring='r2',cv=kf)
print(f'models r2 across 5 cross validations sets : {cv_score}')
print(f'Average r2 from cross validation : {cv_score.mean()}')
print(f'               ----Logistics regresion----\n')
print(f'Train Classification report : \n {classification_report(y_train_smote,y_train_pred)}')
print(f'Test Classification report : \n {classification_report(y_test,y_test_pred)}')

conf_mat = confusion_matrix(y_test,y_test_pred)
print(f'               ----Confusion matrix----\n')
print(conf_mat)
cm_display = ConfusionMatrixDisplay(confusion_matrix=conf_mat)
cm_display.plot(colorbar=False)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python

rfc = RandomForestClassifier(random_state=42)
rfc.fit(X_train_smote,y_train_smote)
kf = KFold(n_splits=5,shuffle=True,random_state=42)
y_train_pred = rfc.predict(X_train_smote)
y_test_pred = rfc.predict(X_test)
y_pred = cross_val_predict(rfc,X_train_smote,y_train_smote,cv=kf)
cv_score = cross_val_score(rfc,X_train_smote,y_train_smote, scoring='r2',cv=kf)
print(f'RFC r2 across 5 cross validations sets : {cv_score}')
print(f'Average r2 from cross validation : {cv_score.mean()}')
print(f'               ----RandomForestClassifier----\n')
print(f'Train Classification report : \n {classification_report(y_train_smote,y_train_pred)}')
print(f'Test Classification report : \n {classification_report(y_test,y_test_pred)}')

conf_mat = confusion_matrix(y_test,y_test_pred)
print(f'               ----Confusion matrix----\n')
print(conf_mat)
cm_display = ConfusionMatrixDisplay(confusion_matrix=conf_mat)
cm_display.plot(colorbar=False)
plt.title('RFC')
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier(random_state=42)
gbc.fit(X_train_smote,y_train_smote)
kf = KFold(n_splits=5,shuffle=True,random_state=42)
y_train_pred = gbc.predict(X_train_smote)
y_test_pred = gbc.predict(X_test)
y_pred = cross_val_predict(gbc,X_train_smote,y_train_smote,cv=kf)
cv_score = cross_val_score(gbc,X_train_smote,y_train_smote, scoring='r2',cv=kf)
print(f'GBC r2 across 5 cross validations sets : {cv_score}')
print(f'Average r2 from cross validation : {cv_score.mean()}')
print(f'               ----GradientBoostingClassifier----\n')
print(f'Train Classification report : \n {classification_report(y_train_smote,y_train_pred)}')
print(f'Test Classification report : \n {classification_report(y_test,y_test_pred)}')

conf_mat = confusion_matrix(y_test,y_test_pred)
print(f'               ----Confusion matrix----\n')
print(conf_mat)
cm_display = ConfusionMatrixDisplay(confusion_matrix=conf_mat)
cm_display.plot(colorbar=False)
plt.show()
```

```python
import pandas as pd

# Get predicted probabilities using the best model (Random Forest here)
y_test_proba = rfc.predict_proba(X_test)[:, 1]  # Probability of turnover

# Create a DataFrame for analysis
results = pd.DataFrame({
    "Employee_ID": X_test.index,  
    "Turnover_Probability": y_test_proba
})

# Categorize into zones
def categorize_risk(prob):
    if prob < 0.20:
        return "Safe Zone (Green)"
    elif prob < 0.60:
        return "Low-Risk Zone (Yellow)"
    elif prob < 0.90:
        return "Medium-Risk Zone (Orange)"
    else:
        return "High-Risk Zone (Red)"

results["Risk_Zone"] = results["Turnover_Probability"].apply(categorize_risk)

# Show distribution
print(results["Risk_Zone"].value_counts())
print('--+'*40)
print(results.head())
```

```python
def model_selection(model_name):

    model = model_name(random_state=42)
    model.fit(X_train_smote,y_train_smote)
    kf = KFold(n_splits=5,shuffle=True,random_state=42)
    y_train_pred = model.predict(X_train_smote)
    y_test_pred = model.predict(X_test)
    y_pred = cross_val_predict(model,X_train_smote,y_train_smote,cv=kf)
    cv_score = cross_val_score(model,X_train_smote,y_train_smote, scoring='r2',cv=kf)
    print(f'{model_name} r2 across 5 cross validations sets : {cv_score}')
    print(f'Average r2 from cross validation : {cv_score.mean()}')
    print(f'               ----{model_name}----\n')
    print(f'Train Classification report : \n {classification_report(y_train_smote,y_train_pred)}')
    print(f'Test Classification report : \n {classification_report(y_test,y_test_pred)}')

    conf_mat = confusion_matrix(y_test,y_test_pred)
    print(f'               ----Confusion matrix----\n')
    print(conf_mat)
    cm_display = ConfusionMatrixDisplay(confusion_matrix=conf_mat)
    cm_display.plot(colorbar=False)
    plt.title(model_name)
    plt.show()
```

```python
model_selection(RandomForestClassifier)
```
